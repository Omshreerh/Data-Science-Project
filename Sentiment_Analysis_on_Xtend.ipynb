```r
# Using the library readr
library(readr)
```

```r
# Reading the CSV and Viewing
amazon <- read_csv("amazon.csv")
View(amazon)
```

```r
# Summary of the dataset
summary(amazon)
```

```r
# Finding the number of rows in columns where value is na
colSums(is.na(amazon))
```

```r
# Removing na elements from dataset and finding the dimension
Amazon.clean<-na.omit(amazon)
```

```r
dim(Amazon.clean)
```

```r
colSums(is.na(Amazon.clean))
# The dataframe now has no NA values
```

```r
library(tidyverse)
glimpse(Amazon.clean)
```

```r
# Finding the unique values in Ratings column
unique(Amazon.clean$Ratings)
```

```r
# Changing the Ratings column to a factor from character variable 
# Factor as they are easier to work with categorical variables
Amazon.clean$Ratings<-as.factor(Amazon.clean$Ratings)
class(Amazon.clean$Ratings)
```

```r
levels(Amazon.clean$Ratings)
```

```r
levels(Amazon.clean$Ratings)[match("1.0 out of 5 stars",levels(Amazon.clean$Ratings))] <- "1"
levels(Amazon.clean$Ratings)[match("2.0 out of 5 stars",levels(Amazon.clean$Ratings))] <- "2"
levels(Amazon.clean$Ratings)[match("3.0 out of 5 stars",levels(Amazon.clean$Ratings))] <- "3"
levels(Amazon.clean$Ratings)[match("4.0 out of 5 stars",levels(Amazon.clean$Ratings))] <- "4"
levels(Amazon.clean$Ratings)[match("5.0 out of 5 stars",levels(Amazon.clean$Ratings))] <- "5"
```

```r
summary(Amazon.clean)
```

```r
library(ggplot2)
ggplot(data.frame(Amazon.clean$Ratings), aes(x=Amazon.clean$Ratings))+geom_bar()
```

```r
counts <- table(Amazon.clean$Ratings)
mosaicplot(counts, xlab='Match Result', ylab='Team',main='Wins by Team', col='orange')
```
